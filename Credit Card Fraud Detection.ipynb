```python
#importing libaries needed 


import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to pandas dataframe
#the first five rows
#reading the CSV file

credit_card_info = pd.read_csv("creditcard.csv")
```

```python
#displaying the csv file 
#the firsst 5 rows
credit_card_info.head()
```

```python
#the last 5 rows
credit_card_info.tail()
```

```python
#information about the dataset
credit_card_info.info()
```

```python
#checks the number of missing values in each column
credit_card_info.isnull().sum()
```

```python
#disterbution of real transactions & fraudulent transactions
credit_card_info['Class'].value_counts()
```

```python
#seperating the data for the analysis

legit = credit_card_info[credit_card_info.Class == 0]
fraud = credit_card_info[credit_card_info.Class == 1]
```

```python
#prints out both transactions
print(legit.shape)
print(fraud.shape)
```

```python
#statistical messures on the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing the lavues for both transactions

credit_card_info.groupby('Class').mean()
```

```python
#taking random data and sampling it
legit_sample = legit.sample(n=492)
```

```python
#concatenating two data frame
new_dataset = pd.concat([legit_sample, fraud],axis=0)
```

```python
#printing the serial number of the values
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
#checking the value counts 
new_dataset['Class'].value_counts()
```

```python
#checks if its agood sample or bad sample
new_dataset.groupby('Class').mean()
```

```python
#spliting the dataset to features and targets 
y = new_dataset['Class']
x = new_dataset.drop(columns = 'Class', axis=1)
```

```python
#printing the labels
print(y)
```

```python
#printing the class column
print(x)
```

```python
#splitting into trainning data & testing data

x_train, x_test, y_train,y_test  = train_test_split(x,y, test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#model trainning 
#LogisticRegression for binary classification problems.

model = LogisticRegression()
```

```python
#trainning the Logistic Regression model with trainning data
#Fits the data into the  LogisticRegression model
model.fit(x_train, y_train)
```

```python
#accuaracy on trainning data

x_train_predection = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_predection,y_train)
```

```python
print('precision on training data:',training_data_accuracy)
```

```python
#accuaracy on test data
x_test_predect = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_predect,y_test)
```

```python
print('precision on testing data:',test_data_accuracy)
```
